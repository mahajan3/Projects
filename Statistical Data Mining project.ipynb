```python
df1=df.drop(['Property2','Property 3','Feature4','Feature11','Feature12','Feature13','Feature14','Feature15','Feature16','Feature20','Feature21','Feature22','Feature23','Feature25'],axis=1)
```

```python
df1
```

```python
from sklearn.preprocessing import StandardScaler
scale= StandardScaler()
df1_x=df1.drop(['Property1'],axis=1)
df1_y=df1['Property1']
```

```python
scaled_data = scale.fit_transform(df1_x)
```

```python
scaled_data = pd.DataFrame(scaled_data, columns=df1_x.columns)
scaled_data
```

```python
scaled_data=scaled_data.join(df1_y)
```

```python
scaled_data
```

```python
from sklearn import preprocessing, svm
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
scaled_data1=scaled_data.tail(80)
```

```python
scaled_data1
```

```python
X=scaled_data1.drop(['Property1'],axis=1)
y=scaled_data1['Property1']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2 , random_state=1)
```

**Linear regression for property1**

```python
linear_regressor = LinearRegression()
linear_regressor.fit(X_train, y_train)
```

```python
y_pred=linear_regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

**Lasso regression for property1**

```python
from sklearn.linear_model import Lasso
model = Lasso(alpha=1.0)
model.fit(X_train, y_train)
```

```python
yhat = model.predict(X_test)
r2_score(y_test,yhat)
```

**Ridge regression for property1**

```python
from sklearn.linear_model import Ridge
model1 = Ridge(alpha=1.0)
model1.fit(X_train, y_train)
yhat1 = model1.predict(X_test)
r2_score(y_test,yhat1)
```

**Predictions for property1**

```python
scaled_data3=scaled_data.head(10)
sx_data=scaled_data3.drop(['Property1'],axis=1)
yhat2=model.predict(sx_data)
yhat2
```
